```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# import seaborn as sns
import statsmodels.api as sm

df1 = pd.read_csv('6_4.txt', sep = ' ', header=0, index_col=0, usecols= [0,1,2,3,4])
df2 = pd.read_csv('6_4.txt', sep = ' ', header=0, index_col=0, usecols= [5,6,7,8,9])
#df = df1.append(df2)
df2.head()
```

```python
df2.columns = list(df1.columns)
df = df1.append(df2)
df
```

```python
pg_fl= df.loc[:,['PGNP','FLFP']]
cm = df.loc[:,'CM']
pg_fl = sm.add_constant(pg_fl)
model = sm.OLS(cm,pg_fl).fit()
model.summary()
```

```python
from scipy.stats import f
```

```python
"{:.100f}".format(f.sf(112,1,62))
```

```python
f = model.f_test('(PGNP = FLFP)')
'{:.2f}'.format(f.pvalue)
```
