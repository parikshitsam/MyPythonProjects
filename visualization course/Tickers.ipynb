```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib notebook
```

# Lets create some Data

```python
y1 = np.random.randn(100)
y2 = np.random.rand(100)
y3 = np.random.randint(0,100,100)
x = np.arange(100)
y4 = 3*(x**2)+2*x+x + 500*y1
```

# Lets do some plotting

```python
fig = plt.figure()
ax1 = plt.subplot(211)
ax1.plot(np.arange(2000, 2010), range(10))

ax1.get_xaxis().get_major_formatter().set_useOffset(True)
#ax1.tick_params(direction='in',length=0.3,width=1,c='blue',labelrotation = 45)
ax2 = plt.subplot(212)
ax2.plot(x,y4, 'r-')
loc = ticker.NullLocator()
ax2.xaxis.set_major_locator(loc)
ax2.yaxis.set_major_locator(loc)
plt.xlim([0,100])
```
