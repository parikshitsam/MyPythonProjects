```python
# Histograms
```

```python
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2, 2, sharex = True)
axs = [ax1, ax2, ax3, ax4]

for n in range(len(axs)):
    samplesize = 10**(n+1)
    sample = np.random.normal(loc= 0.0, scale=1.0, size= samplesize)
    axs[n].hist(sample)
    axs[n].set_title("n = {}".format(samplesize))
```

```python
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2, 2, sharex = True)
axs = [ax1, ax2, ax3, ax4]

for n in range(len(axs)):
    samplesize = 10**(n+1)
    sample = np.random.normal(loc= 0.0, scale=1.0, size= samplesize)
    axs[n].hist(sample, bins=100)
    axs[n].set_title("n = {}".format(samplesize))
```

```python
import matplotlib.gridspec as gridspec

plt.figure()
gspec = gridspec.GridSpec(3,3)

bottom_histogram = plt.subplot(gspec[2,1:])
side_histogram = plt.subplot(gspec[0:2,0])
main_scatter = plt.subplot(gspec[0:2,1:])
```

```python
Y = np.random.normal(loc = 0.0, scale=1.0, size= 10000)
X = np.random.random(size= 10000)
main_scatter.scatter(X, Y, s=1)
bottom_histogram.hist(X, bins=100)
s = side_histogram.hist(Y, bins= 100, orientation= "horizontal")
```

```python
import matplotlib.animation as animation

n= 100
x= np.random.randn(n)
```

```python
def update(curr):
    if curr == n:
        a.event_source.stop()
    plt.cla()
    bins = np.arange(-4, 4, 0.5)
    plt.hist(x[:curr], bins = bins)
    plt.axis([-4, 4, 0, 30])
    plt.gca().set_title("Sampling the Normal Distribution")
    plt.gca().set_ylabel("Frequency")
    plt.gca().set_xlabel("Vaue")
    plt.annotate("n={}".format(curr), [3,27])
```

```python
fig = plt.figure()
a = animation.FuncAnimation(fig, update, interval = 100)
```
